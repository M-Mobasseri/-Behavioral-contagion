```python
import scipy.io as sio
mat = sio.loadmat('C:/Users/Elecomp/Desktop/Lab/Run3Results/resultsROI_Subject012_Condition001.mat')
z = mat['Z']
import numpy as np
# Set diagonal to 0
np.fill_diagonal(z, 0)

# Count number of positive links
num_pos = np.sum(z > 0)

# Count number of negative links
num_neg = np.sum(z < 0)

print("Number of positive links:", num_pos)
print("Number of negative links:", num_neg)
def count_triads(conn):
    n_nodes = conn.shape[0]
    n_balanced = 0
    n_imbalanced = 0

    for i in range(n_nodes):
        for j in range(i+1, n_nodes):
            for k in range(j+1, n_nodes):
                triad = np.sign(conn[i,j]) * np.sign(conn[j,k]) * np.sign(conn[k,i])
                if triad > 0:
                    n_balanced += 1
                elif triad < 0:
                    n_imbalanced += 1

    return n_balanced, n_imbalanced
n_balanced, n_imbalanced = count_triads(z)
print("sub1run1:")
print("Balanced:", n_balanced)
print("Imbalanced:", n_imbalanced)
balance_energy = -(n_balanced - n_imbalanced) /(n_balanced + n_imbalanced)

print("Balance energy:")
print("sub1run1:", balance_energy)
def count_triad_links(conn):

  n_0neg = 0
  n_1neg = 0
  n_2neg = 0
  n_3neg = 0

  for i in range(n_nodes):
    for j in range(i+1, n_nodes):
      for k in range(j+1, n_nodes):

        if i == j or i == k or j == k:
          continue

        n_neg = 0
        if conn[i,j] < 0:
          n_neg += 1
        if conn[j,k] < 0:
          n_neg += 1
        if conn[k,i] < 0:
          n_neg += 1

        if n_neg == 0:
          n_0neg += 1
        elif n_neg == 1:
          n_1neg += 1
        elif n_neg == 2:
          n_2neg += 1
        elif n_neg == 3:
          n_3neg += 1

  return n_0neg, n_1neg, n_2neg, n_3neg
# Number of nodes
n_nodes = z.shape[0]
# Count triad links
n_0neg, n_1neg, n_2neg, n_3neg = count_triad_links(z)

# Print results
print("Triads with 0 neg links:T3=", int(n_0neg))
print("Triads with 1 neg links:T2=", int(n_1neg))
print("Triads with 2 neg links:T1=", int(n_2neg))
print("Triads with 3 neg links:T0=", int(n_3neg))
import numpy as np

def calculate_TMH(conn_matrix):
    # Calculate degrees (weighted degrees)
    degree = np.sum(conn_matrix, axis=1)
    pos_degree = np.sum(np.maximum(conn_matrix, 0), axis=1)  # Positive weighted degree
    neg_degree = np.sum(np.minimum(conn_matrix, 0), axis=1)  # Negative weighted degree

    # Calculate TMH
    TMH = np.sum(degree**2) / np.sum(degree)
    
    # Calculate positive and negative TMH
    posTMH = np.sum(pos_degree**2) / np.sum(np.maximum(degree, 1))  # Ensure denominator is not zero
    negTMH = np.sum(neg_degree**2) / np.sum(np.maximum(degree, 1))  # Ensure denominator is not zero

    output = np.round([TMH, posTMH, negTMH], 3)
    output_names = ['TMH', 'posTMH', 'negTMH']

    return dict(zip(output_names, output))

# Calculate for  condition
tmh = calculate_TMH(z)
# Print the results
print("sub1run1:")
print(tmh)
```
